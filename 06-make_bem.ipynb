```python
def check_skull(subjects_dir, bids_derivatives, subject):
    
    import mne
    import matplotlib
    matplotlib.use('Agg')
    plot_bem_kwargs = dict(
        subject='sub-'+subject,
        subjects_dir=subjects_dir,
        brain_surfaces="white",
        orientation="coronal",
        slices=[50, 100, 150, 200])

    f = mne.viz.plot_bem(**plot_bem_kwargs)

    f.savefig(bids_derivatives + 'sub-' + subject + '/' + 
                           subject + '_plot_bem.png', dpi=150)

def make_bem(subjects_dir, bids_derivatives, subject, ico = 5):

    import mne
    
    conductivity = (0.3,)
    layer = '_single_layer'
    
    model = mne.make_bem_model(subject='sub-'+subject,
                               ico=ico,
                               conductivity=conductivity,
                               subjects_dir=subjects_dir)
    bem = mne.make_bem_solution(model)
    mne.write_bem_solution(bids_derivatives + 'sub-' + subject + '/' + 
                           subject + '_bem_model_ico' + str(ico) + layer, bem, overwrite=True)


subjects_names = ['Z201']

subjects_dir = r'D:\ds005234\derivatives\freesurfer' 
bids_derivatives = bids_root + r'derivatives\\'

for subject in subjects_names:
    check_skull(subjects_dir, bids_derivatives, subject)

for subject in subjects_names:
    make_bem(subjects_dir, bids_derivatives, subject)
```
