```python
def corregistration(subjects_dir, derivatives_folder, subject):
    import mne
    import mne_bids

    bids_path = mne_bids.BIDSPath(subject=subject,
                         root=derivatives_folder,
                         task='vowels',
                         datatype='meg',
                         processing='tsss+mc+transtooptimal+0110hz+ssp',
                         extension='.pos')
                             
    inst = str(bids_path.fpath) + '_meg.fif'
    
    mne.gui.coregistration(subject='sub-'+subject, subjects_dir=subjects_dir, inst=inst)

def check_digit_locations(subjects_dir, derivatives_folder, subject):
    
    import mne
    import mne_bids
    import matplotlib.pyplot as plt
    
    bids_path = mne_bids.BIDSPath(subject=subject,
                         root=derivatives_folder,
                         task='vowels',
                         datatype='meg',
                         processing='tsss+mc+transtooptimal+0110hz+ssp',
                         extension='.pos')
    
    inst = str(bids_path.fpath) + '_meg.fif'
    raw = mne.io.read_raw_fif(inst)
    
    trans = derivatives_folder + 'sub-' + subject + '/meg/sub-' + subject + '_raw-trans.fif'
    
    fig = mne.viz.plot_alignment(
        subject='sub-'+subject,
        info=raw.info,
        trans=trans,
        surfaces="head-dense",
        subjects_dir=subjects_dir,
        meg=False,
        dig=True)

    mne.viz.set_3d_view(fig, azimuth=135, elevation=80, distance=0.4, focalpoint=(0.0, 0.0, 0.0))

    screenshot = fig.plotter.screenshot()

    fig, ax = plt.subplots(figsize=(10, 10))
    ax.imshow(screenshot, origin='upper')
    ax.set_axis_off() 
    fig.tight_layout()
    fig.savefig(derivatives_folder + 'sub-' + subject + '/' +subject + '_plot_alignment.png', dpi=150)
    plt.close('all')

def correct_bem(subjects_dir, subject):
    
    import mne
       
    mne.bem.make_scalp_surfaces(subject, subjects_dir, overwrite=True)
    mne.bem.make_watershed_bem(subject, subjects_dir, overwrite=True)


subjects_dir = r'D:\ds005234\derivatives\freesurfer' 
bids_root = r'D:\\ds005234\\'
bids_derivatives = bids_root + r'derivatives\\'


subject = 'Z201'

trans = r'D:\ds005234\derivatives\all_trans'
corregistration(subjects_dir, bids_derivatives, subject)

```
