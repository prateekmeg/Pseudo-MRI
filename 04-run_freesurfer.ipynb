```python
import os
import subprocess
import shutil

bids_root = r'D:\ds005234'
bids_derivatives = os.path.join(bids_root, 'derivatives')
fs_dir = os.path.join(bids_derivatives, 'freesurfer')

subjects_names = ['Z201']

def run(cmd):
    p = subprocess.Popen(cmd, shell=True, stdout=subprocess.PIPE, stderr=subprocess.PIPE, text=True)
    out, err = p.communicate()
    print(out)
    print(err)
    return p.returncode

for subject in subjects_names:
    subject = 'sub-' + subject
    sdsubject = os.path.join(fs_dir, subject)

    t1w = os.path.join(bids_root, subject, 'anat', subject + '_T1w.nii.gz')
    run(f'recon-all -i "{t1w}" -s {subject}')
    run(f'recon-all -autorecon-all -s {subject}')

    run(f'mne_watershed_bem --subject {subject} --atlas --overwrite')

    iu = os.path.join(sdsubject, 'bem', 'watershed', subject + '_inner_skull_surface')
    ou = os.path.join(sdsubject, 'bem', 'watershed', subject + '_outer_skull_surface')
    oi = os.path.join(sdsubject, 'bem', 'watershed', subject + '_outer_skin_surface')

    shutil.copy(iu, os.path.join(sdsubject, 'bem', 'inner_skull.surf'))
    shutil.copy(ou, os.path.join(sdsubject, 'bem', 'outer_skull.surf'))
    shutil.copy(oi, os.path.join(sdsubject, 'bem', 'outer_skin.surf'))

    run(f'mkheadsurf -subjid {subject}')
    run(f'mne_surf2bem --surf "{sdsubject}\surf\lh.seghead" --id 4 --check --fif "{sdsubject}\bem\{subject}-head-dense.fif"')

    shutil.move(os.path.join(sdsubject, 'bem', f'{subject}-head.fif'),
                os.path.join(sdsubject, 'bem', f'{subject}-head_old.fif'))

    shutil.copy(os.path.join(sdsubject, 'bem', f'{subject}-head-dense.fif'),
                os.path.join(sdsubject, 'bem', f'{subject}-head.fif'))
```
